# Relationship between data

```python
dfs = {
    'device': device_df, 
    'email': email_df, 
    'file': file_df, 
    'http': http_df, 
    'logon': logon_df, 
    'psych': psycho_df
}
```

```python
names = [k for k in dfs]
cols = list(c for df in dfs.values() for c in df.columns)

cols = [c for i, c in enumerate(cols) if c not in cols[:i]] # filter duplicates, preserve order
```

```python
rel_df = pd.DataFrame(index=names, columns=cols)
rel_df = rel_df.fillna(False)
for name, df in dfs.items():
    for c in df.columns:
        rel_df.loc[name, c] = True
        
rel_df
```

## User IDs

Looking at intersections of User IDs in the datasets - whether they relate to the same user IDs or not?
But based on just samples of data so not correct..

```python
uids_df = pd.DataFrame(index=names, columns=names)
```

```python
file_df.user.nunique()
```

```python
for k1, df1 in dfs.items():
    for k2, df2 in dfs.items():
        inter = len(set(df1['user']).intersection(set(df2['user'])))
        uids_df.loc[k1, k2] = inter / len(set(df1['user']))
```
